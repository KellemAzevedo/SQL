```python
'''
Agora, vamos a um exemplo mais complexo, em que desejamos visualizar o nome do
cliente e o modelo do carro alugado. Nesse caso, vamos precisar realizar dois comandos JOINS.
'''
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
)
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
SELECT cl.nome, ca.modelo
FROM cliente AS cl
JOIN aluguel AS al
 ON cl.codcliente = al.codcliente
JOIN carro AS ca
 ON al.codcarro = ca.codcarro;
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
result = cursor.fetchall() # Recuperação de todos os resultados da consulta
pedidos = pd.DataFrame(result, columns=cursor.column_names) # Criação de um DataFrame Pandas usando os resultados da consulta e os nomes das colunas do cursor
display(pedidos) # Exibição do DataFrame Pandas contendo os resultados da consulta
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

```python
'''
EXEMPLO
Para dar mais um exemplo, vamos recuperar, em nosso banco de dados, a lista de todos
os clientes e a data dos aluguéis realizados por eles. Porém, queremos, também, como
resultado aqueles clientes que não alugaram nenhum carro.
O resultado da nossa consulta fica:
'''
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
)
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
SELECT cl.nome, al.dataaluguel
FROM cliente AS cl
LEFT JOIN aluguel AS al
 ON cl.codcliente = al.codcliente;
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
result = cursor.fetchall() # Recuperação de todos os resultados da consulta
pedidos = pd.DataFrame(result, columns=cursor.column_names) # Criação de um DataFrame Pandas usando os resultados da consulta e os nomes das colunas do cursor
display(pedidos) # Exibição do DataFrame Pandas contendo os resultados da consulta
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

```python
# Right Join
'''
EXEMPLO
Para dar mais um exemplo, vamos recuperar, em nosso banco de dados, a lista de todos
os clientes e a data dos aluguéis realizados por eles. Porém, queremos, também, como
resultado aqueles clientes que não alugaram nenhum carro.
O resultado da nossa consulta fica:
'''
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
)
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
SELECT cl.nome, al.dataaluguel
FROM aluguel AS al
RIGHT JOIN cliente AS cl
 ON cl.codcliente = al.codcliente;
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
result = cursor.fetchall() # Recuperação de todos os resultados da consulta
pedidos = pd.DataFrame(result, columns=cursor.column_names) # Criação de um DataFrame Pandas usando os resultados da consulta e os nomes das colunas do cursor
display(pedidos) # Exibição do DataFrame Pandas contendo os resultados da consulta
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

```python
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
)
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
SELECT *
  FROM      carro AS ca
  LEFT JOIN aluguel al ON al.codcarro = ca.codcarro
UNION
SELECT *
  FROM      carro AS ca
  RIGHT JOIN aluguel al ON al.codcarro = ca.codcarro;
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
result = cursor.fetchall() # Recuperação de todos os resultados da consulta
pedidos = pd.DataFrame(result, columns=cursor.column_names) # Criação de um DataFrame Pandas usando os resultados da consulta e os nomes das colunas do cursor
display(pedidos) # Exibição do DataFrame Pandas contendo os resultados da consulta
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```
